```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds

def encode_time(ds, reference_time = '1958-01-01', period = 'months'):
    ds.time.encoding['calendar'] = '360_day'
    ds.time.encoding['units'] = '{} since {}'.format(period, reference_time)
    return ds

```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# Assuming your data arrays are named 'w' and 'sst'
# Load the data
# ds = xr.open_dataset('your_data.nc')


# Define SST bins
sst_bins = np.arange(0, 31, 2)

# Bin the SST data
sst_binned = np.digitize(sst['sst_dashed'], sst_bins)

# Create DataArray with bin indices as coordinate
sst_binned_da = xr.DataArray(
    sst_binned,
    coords=sst['sst_dashed'].coords,
    dims=sst['sst_dashed'].dims,
    name='sst_binned'
)

# Compute mean wind speed for each SST bin
mean_winds = sst_binned_da.groupby_bins(sst_binned_da, sst_bins).mean()

# Plotting
plt.figure(figsize=(10, 6))
mean_winds.plot(marker='o', color='b')

plt.xlabel('SST (°C)')
plt.ylabel('Mean Wind Speed (m/s)')
plt.title('Mean Wind Speed for SST Bins')

plt.grid(True)
plt.tight_layout()
plt.show()
```

# Linear Trend

```python

from scipy.stats import linregress


ds = sst.copy()
# Assuming your xarray dataset is named 'ds' and has dimensions ('time', 'lat', 'lon')
# You should replace this with your actual dataset

# Calculate monthly climatology
climatology = ds.sst.groupby('time.month').mean(dim='time')

# Compute anomalies
anomalies = ds.sst.groupby('time.month') - climatology

# Reshape anomalies to 2D array (time, space)
anomalies_2d = anomalies.stack(space=('lat', 'lon'))

# Calculate linear trend per decade and r-value
def trend_per_decade(y):
    x = np.arange(len(y))
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    trend_per_decade = slope * 10  # Multiply by 10 to get trend per decade
    return trend_per_decade, r_value

# Apply the function along the time dimension
trend_per_decade, r_value = np.apply_along_axis(trend_per_decade, axis=0, arr=anomalies_2d)

# Reshape the arrays back to the original shape
trend_per_decade = trend_per_decade.reshape(ds.dims['lat'], ds.dims['lon'])
r_value = r_value.reshape(ds.dims['lat'], ds.dims['lon'])

# Create xarray DataArrays for the trend per decade and r-value
trend_per_decade_da = xr.DataArray(trend_per_decade, dims=('lat', 'lon'), coords={'lat': ds.lat, 'lon': ds.lon})
r_value_da = xr.DataArray(r_value, dims=('lat', 'lon'), coords={'lat': ds.lat, 'lon': ds.lon})

# Assign attributes
trend_per_decade_da.attrs['units'] = '°C/decade'
r_value_da.attrs['description'] = 'Correlation coefficient of linear regression'

# Optional: Replace NaN values with fill value if needed
# trend_per_decade_da = trend_per_decade_da.fillna(fill_value)
# r_value_da = r_value_da.fillna(fill_value)

# Optional: Merge DataArrays into a single Dataset if desired
# trend_and_r_value_ds = xr.Dataset({'trend_per_decade': trend_per_decade_da, 'r_value': r_value_da})

# Save the DataArrays or Dataset to NetCDF file if desired
# trend_and_r_value_ds.to_netcdf('trend_and_r_value.nc')

# Print or further process the results as needed
print(trend_per_decade_da)
print(r_value_da)
```

```python
fp = '/data/jmex/ERA5/monthly/era5_tp_anom_mthly.nc'
tp = decode_time(fp).sel(lat = slice(20,-20), lon = (130,150))
nino = tp.tp.rolling(time = 3).mean().sel(time = '2023')
nino2 = tp.tp.rolling(time = 3).mean().sel(time = '2015')
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

# Generate sample Nino3.4 data (replace this with your own data loading)
# Sample data for demonstration

# Define El Niño and La Niña thresholds (you may adjust these)
el_nino_threshold = 0.5
la_nina_threshold = -0.5

# Initialize figure and axis
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Define colors for El Niño and La Niña
cmap = plt.get_cmap('coolwarm')
norm = Normalize(vmin=min(nino.min(), la_nina_threshold), vmax=max(nino.max(), el_nino_threshold))
scalar_map = ScalarMappable(norm=norm, cmap=cmap)

# Plot monthly values for all years individually
months = np.arange(1, 13)
for month in months:
    monthly_data = nino.sel(time=nino['time.month'] == month).mean()
    theta =  2 * np.pi * (month-1)/12
    r = monthly_data.values
    color = scalar_map.to_rgba(r)  # Color based on anomaly value
    ax.plot(theta, r, marker='o', linestyle = '--', markersize=5, label=f'Month {month}', color='b')


# Plot monthly values for all years individually
months = np.arange(1, 13)
for month in months:
    monthly_data = nino2.sel(time=nino['time.month'] == month).mean()
    theta =  2 * np.pi * (month-1)/12
    r = monthly_data.values
    color = scalar_map.to_rgba(r)  # Color based on anomaly value
    ax.plot(theta, r, marker='o', linestyle = '--', markersize=5, label=f'Month {month}', color='r')

# Add legend
#ax.legend(loc='upper right')

# Show plot
plt.title('Nino3.4 Index Anomaly Clock Plot')
plt.show()
```
