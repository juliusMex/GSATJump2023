```python
# the libraries
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import cartopy.crs as ccrs


from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds

def encode_time(ds, reference_time = '1958-01-01', period = 'months'):
    ds.time.encoding['calendar'] = '360_day'
    ds.time.encoding['units'] = '{} since {}'.format(period, reference_time)
    return ds


def build_weights(data, datadata, minvalue = -999, latlim = 90.):
    weights = np.cos( np.deg2rad(data.lat))
    weights = weights*(abs(data.lat)<latlim)
    return weights

def weighted(data, datadata, weights):
    datadata_weighted = datadata.weighted(weights)
    weighted_mean = datadata_weighted.mean(('lat', 'lon'))
    return weighted_mean

def box_sum(data, datadata, weights):
    datadata_weighted = datadata.weighted(weights)
    weighted_sum = datadata_weighted.sum(('lat', 'lon'))
    return weighted_sum

def spaghetti(dataset, index_name, outputname, ylabel = ''):
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(12, 6))

    # Use Seaborn's lineplot to create the spaghetti plot
    sns.lineplot(data = dataset, x = dataset.time.dt.month, y = index_name,
                 hue = dataset.time.dt.year, palette=['lightgrey' for y in np.unique(dataset.time.dt.year)], style=None, ax=ax, legend=False)

    # Manually select years to highlight with different colors
    vstrong_ninos = [1982, 1997, 2015]
    strong_ninos = [1965, 1972,1987, 1991]

    #plot 2023
    sns.lineplot(data=dataset.where(dataset.time.dt.year == 2023), x=dataset.time.dt.month, y=index_name, color='blue', ax=ax, label=str(2023))

    for year in vstrong_ninos:
        sns.lineplot(data=dataset.where(dataset.time.dt.year == year), x=dataset.time.dt.month, y=index_name, color='red', ax=ax, label=str(year))

    for year in strong_ninos:
        sns.lineplot(data=dataset.where(dataset.time.dt.year == year), x=dataset.time.dt.month, y=index_name, color='orange', ax=ax, label=str(year))

    # Customize the plot
    ax.set_xlabel('Month')
    ax.set_ylabel(ylabel)
    ax.set_title(index_name)
    ax.legend(loc = 'lower right')

    # Show the plot
    plt.savefig('/data/jmex/output/figures/{}.png'.format(outputname))
```

```python
months = ['DJF', 'JFM', 'FMA','MAM', 'AMJ', 'MJJ', 'JJA', 'JAS',  'ASO', 'SON', 'OND', 'NDJ']
```

```python
fp = '/data/jmex/ERA5/monthly/era5_tp.nc'
tp = decode_time(fp)#select rev. time only
#select revtime only
tp = tp.sel(time = slice('1958-01-01', '2024-01-31'))
#select all nino area
#tp = tp.sel(lat = slice(15,-15), lon = slice(110, 270))
```

```python
# taking the rollig mean instead
tp = tp.rolling(time = 3).mean()
```

```python
vstrong = [1982, 1997, 2015]
linestyles = ['-', '--', '-.', ':']

strong = [1965, 1972, 1987, 1991]

tp_vstrong = tp.where(tp.time.dt.year.isin(vstrong), drop = True)
tp_strong = tp.where(tp.time.dt.year.isin(strong), drop = True)
```

# Zonal, individual Lines

Compute the zonal averages.
1. Rolling Mean of 3Months
2. Vstrong ENSO individual lines with labels
3. Strong ENSO oragne lines no labels
4. greyshaded 10-890th percentile of all values

```python
for month in range(0,12):
    fig, ax = plt.subplots(figsize = (10,6))
    
    #plot mean and std for given month
    m = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = ['time', 'lon'])
    
    
    #compute 10 and 90th percentile
    perc10 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = 'lon') \
    .quantile(0.1, dim = 'time') 
    
    perc90 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = 'lon') \
    .quantile(0.9, dim = 'time')
    
    #m.tp.plot(ax =ax, label = 'Mean', color = 'k')
    ax.fill_between(perc10.lat, perc10.tp , perc90.tp, color = 'lightgrey',
                    alpha = 1, label = '10th - 90th Percentile')
    
    
    #add very storng and strong ENSO lines
    for i, y in enumerate(vstrong): 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.mean(dim = 'lon')\
        .plot(ax = ax, color = 'r', linestyle = linestyles[i], label = 'Very Strong ENSO '+str(y))
    
    for y in strong: 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.mean(dim = 'lon').plot(ax = ax, color = 'orange')
    #add a single legend entry of strong ENSO
    plt.plot([], [], linestyle='-', color='orange', label='Strong ENSOs ')

    
    
    #plot the year 2023
    tp.sel(time = '2023-{}'.format(month+1)).mean(dim = 'lon').tp.plot(ax= ax, color = 'b', label = '2023')
    
    
    # Customize the plot
    
    ax.set_ylim(0,12)

    ax.set_title('Zonal Mean Precipitation {} '.format(months[month]))
    ax.set_xlabel('Latitude [deg]')
    ax.set_ylabel('Precipitation [mm/day]')
    ax.legend(loc = 'upper right')
    plt.savefig('/data/jmex/output/figures/precip/zonal/mean/zonal_mean_{}.png'.format(month+1))
```

```python
for month in range(0,12):
    fig, ax = plt.subplots(figsize = (10,6))
    
    #plot mean and std for given month
    m = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .weighted(build_weights(tp, tp.tp))\
    .mean(dim = ['time', 'lat'])
    
    perc10 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .weighted(build_weights(tp, tp.tp))\
    .mean(dim = 'lat') \
    .quantile(0.1, dim = 'time') 
    
    perc90 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .weighted(build_weights(tp, tp.tp))\
    .mean(dim = 'lat') \
    .quantile(0.9, dim = 'time') 
    
    
    #m.tp.plot(ax =ax, label = 'Mean', color = 'k')
    ax.fill_between(perc10.lon, perc10.tp, perc90.tp, color = 'lightgrey',
                    alpha = 0.9, label = '10th-90th Percentile')
    
    for y in vstrong: 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.weighted(build_weights(tp, tp.tp))\
        .mean(dim = 'lat')\
        .plot(ax = ax, color = 'r', linestyle = linestyles[i], label = 'Very Strong ENSO '+str(y))
        
    
    for y in strong: 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.weighted(build_weights(tp, tp.tp))\
        .mean(dim = 'lat').plot(ax = ax, color = 'orange')
    #add a single legend entry of strong ENSO
    plt.plot([], [], linestyle='-', color='orange', label='Strong ENSOs ')
    
    
  
    #plot the year 2023
    tp.sel(time = '2023-{}'.format(month+1))\
    .weighted(build_weights(tp, tp.tp))\
    .mean(dim = 'lat').tp.plot(ax= ax, color = 'b', label = '2023')
    
    
    # Customize the plot
    ax.set_ylim(1,6)
    
    ax.set_title('Meridional Mean Precipitation {}'.format(months[month]))
    ax.set_xlabel('Longitude [deg]')
    ax.set_ylabel('Precipitation [mm/day]')
    ax.legend(loc = 'upper left')
    
    #set box boundaries
    ax.axvline(x = 130,linestyle = '--', color = 'grey')
    ax.axvline(x = 160,linestyle = '--', color = 'grey')
    ax.axvline(x = 210,linestyle = '--', color = 'grey')
    ax.axvline(x = 270,linestyle = '--', color = 'grey')
    ax.axvline(x = 280,linestyle = '--', color = 'grey')
    
    #set text
    ax.text(135, 1.1, s = 'Nino West', rotation = 90)
    ax.text(165, 1.1, s = 'Nino 4', rotation = 90)
    ax.text(215, 1.1, s = 'Nino 3', rotation = 90)
    ax.text(275, 1.1, s = 'Nino 1+2', rotation = 90)
    
    
    
    
    
    plt.savefig('/data/jmex/output/figures/precip/meridional/mean/meridional_mean_{}.png'.format(month+1))
    
```

# Anomalies


How anomalies where computed:

1. #compute climatology for each field
 climatology = tp.groupby('time.month').mean('time')
2. #substract climatology for each field
danom = tp.groupby('time.month')-climatology
3. #normalise for each field
danom = danom/danom.std(dim = 'time')
danom = danom.rename({'tp':'tp_anom'})
4. Save
danom.to_netcdf('/data/jmex/ERA5/monthly/era5_tp_anom_mthly.nc')

```python
fp = '/data/jmex/ERA5/monthly/era5_tp.nc'
tp = decode_time(fp)#select rev. time only
#select revtime only
tp = tp.sel(time = slice('1958-01-01', '2024-01-31'))
#compute climatology for each field
climatology = tp.groupby('time.month').mean('time')
#substract climatology for each field
danom = tp.groupby('time.month')-climatology


danom.to_netcdf('/data/jmex/ERA5/monthly/era5_tp_anom_mthly.nc')
```

Compute the zonal averages.

3. Strong ENSO oragne lines no labels
4. greyshaded 10-890th percentile of all values

```python
for month in range(9,10):
    fig, ax = plt.subplots(figsize = (10,6))
    
    #plot mean and std for given month
    m = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = ['time', 'lon'])
    
    
    #compute 10 and 90th percentile
    perc10 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = 'lon') \
    .quantile(0.1, dim = 'time') 
    
    perc90 = tp.where(tp.time.dt.month == (month+1), drop = True)\
    .mean(dim = 'lon') \
    .quantile(0.9, dim = 'time')
    
    m.tp.plot(ax =ax, label = 'Mean', color = 'k')
    ax.fill_between(perc10.lat, perc10.tp , perc90.tp, color = 'lightgrey',
                    alpha = 1, label = '10th - 90th Percentile')
    
    
    #add very storng and strong ENSO lines
    for i, y in enumerate(vstrong): 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.mean(dim = 'lon')\
        .plot(ax = ax, color = 'r')
    plt.plot([], [], linestyle='-', color='red', label='Very Strong ENSOs ')

    
    for y in strong: 
        tp.sel(time = '{}-{}'.format(y, month+1)).tp.mean(dim = 'lon').plot(ax = ax, color = 'orange')
    #add a single legend entry of strong ENSO
    plt.plot([], [], linestyle='-', color='orange', label='Strong ENSOs ')

    
    
    #plot the year 2023
    tp.sel(time = '2023-{}'.format(month+1)).mean(dim = 'lon').tp.plot(ax= ax, color = 'b', label = '2023')
    
    
    # Customize the plot
    
    ax.set_ylim(-2,4)

    ax.set_title('Zonal Mean of Precipitation {} '.format(months[month]))
    ax.set_xlabel('Latitude [deg]')
    ax.set_ylabel('Anomaly [mm/day]')
    ax.legend(loc = 'upper left')
    #plt.savefig('/data/jmex/output/figures/precip/zonal/anomaly/zonal_anomaly_{}.png'.format(month+1))
```
