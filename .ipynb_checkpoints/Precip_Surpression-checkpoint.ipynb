```python
# the libraries
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy


from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds

def encode_time(ds, reference_time = '1958-01-01', period = 'days'):
    ds.time.encoding['calendar'] = '360_day'
    ds.time.encoding['units'] = '{} since {}'.format(period, reference_time)
    return ds


def build_weights(data, datadata, minvalue = -999, latlim = 90.):
    weights = np.cos( np.deg2rad(data.lat))
    weights = weights*(abs(data.lat)<latlim)
    return weights

def anomaly(glbmean, normalise = False, detrend = False, normalise_by = 'time.month'):
    
    #nanoseconds per century required for fit
    ns_century = 1.e9*3600.*24.*365.*100.                # scaling: nono-seconds to century
    #detrend if desired
    if detrend:
        linfit = glbmean.polyfit('time',1)
        slope = linfit.polyfit_coefficients.isel(degree = 0)*ns_century

        trend = xr.polyval(coord = glbmean.time, coeffs = linfit.polyfit_coefficients)
        glbmean = glbmean - trend.values +glbmean.mean(dim = 'time')
    
    
    databymth = glbmean.groupby(normalise_by)
    mthclim = databymth.mean("time")       # detrended climatological months (seasonal cycle)
    
    dataanom = databymth - mthclim    # detrended interannual annomaly
   

    if normalise:
        dataanom = dataanom/dataanom.std()
    dataanom = dataanom.rename('anom')   # change variable name

    return dataanom

def weighted(data, datadata, weights):
    datadata_weighted = datadata.weighted(weights)
    weighted_mean = datadata_weighted.mean(('lat', 'lon'))
    return weighted_mean
```

```python
def spaghetti(dataset, index_name, outputname, title = ''):
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(12, 6))

    # Use Seaborn's lineplot to create the spaghetti plot
    sns.lineplot(data = dataset, x = dataset.time.dt.dayofyear, y = index_name,
                 hue = dataset.time.dt.year,
                 palette=['lightgrey' for y in np.unique(dataset.time.dt.year)],
                 style=None,
                 ax=ax,
                 legend=False)

    # Manually select years to highlight with different colors
    vstrong_ninos = [1982, 1997, 2015]
    styles = []
    #strong_ninos = [1965, 1972,1987, 1991]

    #plot 2023
    sns.lineplot(data=dataset.where(dataset.time.dt.year == 2023), x=dataset.time.dt.month, y=index_name, color='blue', ax=ax, label=str(2023))

    sns.lineplot(data=dataset.where(dataset.time.dt.year == 2015), x=dataset.time.dt.month, y=index_name, color='red', ax=ax, label = '2015')
    sns.lineplot(data=dataset.where(dataset.time.dt.year == 1997), x=dataset.time.dt.month, y=index_name, color='darkorange', ax=ax, label = '1997')
    sns.lineplot(data=dataset.where(dataset.time.dt.year == 1982), x=dataset.time.dt.month, y=index_name, color='orange', ax=ax, label = '1982')

    #for year in strong_ninos:
     #   sns.lineplot(data=dataset.where(dataset.time.dt.year == year), x=dataset.time.dt.dayofyear, y=index_name, color='orange', ax=ax, label=str(year))

    # Customize the plot
    ax.set_xlabel('Month')
    ax.set_ylabel(index_name)
    ax.set_title(title)
    ax.legend(loc = 'lower right')

    # Show the plot
    #plt.savefig('/data/jmex/output/figures/same_start/{}.png'.format(outputname))
```

# Cumulutative Daily Precip

```python
#read in sst
fp = '/data/jmex/ERA5/daily/era5_tp_daily_130-150E_-20-20N.nc'

#load dataset
ds = xr.open_dataset(fp)
#compute weighted average of area
tp = ds.weighted(np.cos(np.deg2rad(ds.lat))).mean(['lat','lon'])

#take 90D rolling mean
tp = tp.rolling(time = 90, center = True).mean()

#take time selection
tp = tp.sel(time = slice('1958', '2023'))

#compute daily anomaly of weighted average
tp_dgroup = tp.groupby('time.dayofyear')
dclim = tp_dgroup.mean('time')
tp_danom = tp_dgroup-dclim

```

```python
strong_nino = [1965, 1972,1987, 1991]

vstrong_nino = [1982, 1997, 2015]

vstrong_nina = [1973, 1975, 1988, 1998, 1999, 2007, 2010]
```

```python
ds= tp_danom

fig, ax = plt.subplots(1,1, figsize = (12,8))

ax.set_xlabel('Day of Year')
ax.set_ylabel('Cumulative Precipitation  Anomaly')
ax.set_title('West Pacific S20N20|E130E150')

ax.axhline(0, linestyle = '--', c = 'k')

for y in np.unique(ds.time.dt.year):
    ax.plot(np.cumsum((ds.sel(time = str(y)).tp.data)), c = 'lightgrey')
    
ax.plot(np.cumsum((ds.sel(time = '2023').tp.data)), c = 'k', label = '2023')

for v in vstrong_nino:
    ax.plot(np.cumsum((ds.sel(time = str(v)).tp.data)), c = 'red', label = v)
    
for s in strong_nino:
    ax.plot(np.cumsum((ds.sel(time = str(s)).tp.data)), c = 'orange', label = s)

#for va in vstrong_nina:
 #   ax.plot(np.cumsum((ds.sel(time = str(va+1)).tp.data)), c = 'lightblue', label = va)

    
ax.legend()

fig.savefig('/data/jmex/output/figures/sandrine/cum_west.png')
```

# West: Daily data

```python
#read in sst
fp = '/data/jmex/ERA5/daily/era5_tp_daily_130-150E_-20-20N.nc'

#load dataset
ds = xr.open_dataset(fp)
#compute weighted average of area
tp = ds.weighted(np.cos(np.deg2rad(ds.lat))).mean(['lat','lon'])

#take 90D rolling mean
tp = tp.rolling(time = 90, center = True).mean()

#take time selection
tp = tp.sel(time = slice('1958', '2023'))

#compute daily anomaly of weighted average
tp_dgroup = tp.groupby('time.dayofyear')
dclim = tp_dgroup.mean('time')
tp_danom = tp_dgroup-dclim

```

```python
# nino index and previous year
fp = '/data/jmex/output/drivers/indices/nino.nc'
nino = decode_time(fp)
ninoOND = nino.sel(time = (nino['time.month']>=10)).groupby('time.year').mean(dim = 'time')

ninoOND_previous = ninoOND.shift(year = 1)
```

## COmpute correlation between Nino34 and 90D-rolling precipitation 

```python

# Calculate Linear Regression
rvalues = []
slopes = []
pvalues = []

# Iterate over each day of the year
for day in range(1, 366):
    # Select data for this day of the year
    tp_values = tp_danom.sel(time=tp_danom['time.dayofyear'] == day)
    
    # Group by year and calculate mean precipitation for each day
    tp_values_grouped = tp_values.groupby('time.year').mean(dim='time')
    
    # Align nino34 values for the corresponding years
    nino_values = ninoOND_previous.sel(year=tp_values_grouped['year'])
    
    # linear regression
    linregress = scipy.stats.linregress(ninoOND.nino34[1:],tp_values_grouped.tp[1:])
    rvalues.append(linregress.rvalue)
    slopes.append(linregress.slope)
    pvalues.append(linregress.pvalue)
    

# Plot the correlations
plt.figure(figsize=(10, 6))
plt.plot(range(1, 366), rvalues, marker='o', linestyle='-')
plt.title('Linear Reg Precipitation vs Nino34 Index by Day of Year')
plt.xlabel('Day of Year')
plt.ylabel('Rvalue')
plt.grid(True)
plt.savefig('/data/jmex/output/figures/precip_nino/lagged/west_rvalue.png')

plt.show()


# Plot the correlations
plt.figure(figsize=(10, 6))
plt.plot(range(1, 366), slopes, marker='o', linestyle='-')
plt.title('Linear Reg Precipitation vs Nino34 Index by Day of Year')
plt.xlabel('Day of Year')
plt.ylabel('Slope [mm/day / K]')
plt.grid(True)

plt.savefig('/data/jmex/output/figures/precip_nino/lagged/west_slopes.png')
plt.show()

```

```python
def scatter(month):
    months = ['DJF', 'JFM', 'FMA', 'MAM', 'AMJ', 'MJJ', 'JJA', 'JAS', 'ASO', 'SON', 'OND', 'NDJ']
    # Select data for the 28th of April
    tp_values = tp_danom.tp.sel(time=(tp_danom.tp['time.month'] == month+1) & (tp_danom.tp['time.day'] == 15))

    # Perform linear regression using scipy.stats.linregress
    slope, intercept, rvalue, pvalue, stderr = scipy.stats.linregress(ninoOND.nino34, tp_values)


    # Define colors based on the previous year's nino34 index
    colors = np.where(ninoOND_previous.nino34 > 0.5, 'red', np.where(ninoOND_previous.nino34 < -0.5, 'blue', 'white'))

    # Plot scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(ninoOND.nino34, tp_values, c=colors, alpha=0.7, edgecolor = 'k', s = 90)
    plt.scatter(ninoOND.nino34.sel(year = 2023), tp_values.isel(time = -1), s = 100, facecolor='none', alpha=1, edgecolor = 'k', linewidth = 4)
    
    for j in range(len(ninoOND.nino34) - 1):
        dx = (ninoOND.nino34[j + 1] - ninoOND.nino34[j])/10.
        dy = (tp_values[j + 1] - tp_values[j])/10.
        plt.arrow(ninoOND.nino34[j], tp_values[j], dx, dy, head_width=0.03, head_length=0.05, fc='k', ec='k')

    
    plt.title('{} Precipitation vs. Nino34 OND'.format(months[i]))
    plt.ylabel('Precipitation Anomaly [mm/day]')
    plt.xlabel('Nino34 Index [K]')

    # Plot regression line
    plt.plot(ninoOND.nino34, slope * ninoOND.nino34 + intercept, color='black')

    # Add slope and rvalue to the plot
    plt.text(0.1, 0.1, f'Slope: {slope:.2f}\nRvalue: {rvalue:.2f}', transform=plt.gca().transAxes)

    plt.grid(True)
    plt.savefig('/data/jmex/output/figures/precip_nino/scatter/west{}_{}.png'.format(i+1,months[i]))
    
    plt.show()
    return None
```
