```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy.stats import moment
from scipy import stats
import matplotlib.colors as mcolors

from scipy.stats import linregress

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns
from matplotlib.cm import ScalarMappable


from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds


# Define constants
p0 = 1000  # reference pressure in hPa
R = 287.05  # specific gas constant for dry air in J/(kg*K)
cp = 1005  # specific heat at constant pressure for dry air in J/(kg*K)

def convert_to_potential_temperature(T, p):
    """
    Convert temperature to potential temperature.
    
    Parameters:
    T (xarray.DataArray): Temperature in K
    p (xarray.DataArray): Pressure in hPa
    
    Returns:
    xarray.DataArray: Potential temperature in K
    """
    return T * (p0 / p) ** (R / cp)
```

```python
import xarray as xr
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculate the correlation map using xr.corr
correlation_map = xr.corr(residuals, sst_aso.sst, dim='year')

# Initialize an array to hold p-values
p_values = np.empty_like(correlation_map.values) * np.nan

# Calculate p-values for each grid point
for i in range(sst_aso.sst.shape[1]):  # lat dimension
    for j in range(sst_aso.sst.shape[2]):  # lon dimension
        y = sst_aso.sst.isel(lat=i, lon=j).values
        x = residuals.values
        
        # Calculate Pearson correlation and p-value if data is present
        if np.all(np.isfinite(y)):
            _, p_value = stats.pearsonr(x, y)
            p_values[i, j] = p_value

# Create an xarray DataArray for p-values
p_values_da = xr.DataArray(p_values, coords=[sst_aso.lat, sst_aso.lon], dims=["lat", "lon"])

# Define significance level and create a mask
alpha = 0.01
significant_mask = (p_values_da < alpha).astype(int)  # Mask as 1 (significant) and 0 (non-significant)

# Identify regions where 10x10 grid cells are all significant
# Use coarsening (downsampling) to check blocks of 10x10
coarse_significant_mask = significant_mask.coarsen(lat=5, lon=5, boundary="trim").mean()
stippling_mask = (coarse_significant_mask == 1)  # True where all 10x10 cells are significant

# Extract lat/lon points for stippling based on the stippling_mask
lat_coarse = coarse_significant_mask.lat
lon_coarse = coarse_significant_mask.lon
stippling_lat, stippling_lon = np.meshgrid(lat_coarse, lon_coarse, indexing="ij")
stippling_points = stippling_mask.values  # Boolean mask for stippling points
```

```python
import xarray as xr
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculate the correlation map using xr.corr
correlation_map = xr.corr(residuals, sst_aso.sst, dim='year')

# Initialize an array to hold p-values
p_values = np.empty_like(correlation_map.values) * np.nan

# Calculate p-values for each grid point
for i in range(sst_aso.sst.shape[1]):  # lat dimension
    for j in range(sst_aso.sst.shape[2]):  # lon dimension
        y = sst_aso.sst.isel(lat=i, lon=j).values
        x = residuals.values
        
        # Calculate Pearson correlation and p-value if data is present
        if np.all(np.isfinite(y)):
            _, p_value = stats.pearsonr(x, y)
            p_values[i, j] = p_value

# Create an xarray DataArray for p-values
p_values_da = xr.DataArray(p_values, coords=[sst_aso.lat, sst_aso.lon], dims=["lat", "lon"])

# Define significance level and create a mask
alpha = 0.01
significant_mask = (p_values_da < alpha).astype(int)  # Mask as 1 (significant) and 0 (non-significant)

# Identify regions where 10x10 grid cells are all significant
# Use coarsening (downsampling) to check blocks of 1x1
coarse_significant_mask = significant_mask.coarsen(lat=5, lon=5, boundary="trim").mean()
stippling_mask = (coarse_significant_mask == 1)  # True where all 10x10 cells are significant

# Extract lat/lon points for stippling based on the stippling_mask
lat_coarse = coarse_significant_mask.lat
lon_coarse = coarse_significant_mask.lon
stippling_lat, stippling_lon = np.meshgrid(lat_coarse, lon_coarse, indexing="ij")
stippling_points = stippling_mask.values  # Boolean mask for stippling points
```
