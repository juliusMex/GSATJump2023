```python
import numpy as np
import xarray as xr
import pandas as pd

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns

from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

```python
#load precip GPCP
fp = '/data/jmex/NOAA/precip.mon.mean.nc'
tp = xr.open_dataset(fp).sel(lat = slice(-20,20), time = slice('1990', '2024'))
tp = tp.drop_vars(['time_bnds', 'lat_bnds', 'lon_bnds']).rename({'precip':'tp'})

#load sst and coarsen

fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20, -20))


#interp sst to match gpcp
sst = sst.interp(lat = tp.lat, lon = tp.lon, method = 'nearest')


#make ocean mask
ocean_mask = ~sst.isel(time = 1).sst.isnull()

#take t for ocean mask only: 
tp = tp.where(ocean_mask)
```

```python
#weights
weights = np.cos(np.deg2rad(sst.lat))

#tp mean tropical (20-20)
tp_mean = tp.weighted(weights).mean(['lat', 'lon'])

#hadamard product of tp and sst
sst['sst_tp'] = sst.sst*tp.tp


#weigth by area
#sst['sst_tp'] = sst.sst_tp.weighted(weights).mean(['lat','lon'])

#normalise by tp mean
sst['sst_tp'] = sst.sst_tp/tp.tp

#anomaly
sst['sst_tp_anom'] = sst.sst_tp.groupby('time.month') - sst.sst_tp.groupby('time.month').mean()

sst['sst_tp_anom'] = sst.sst_tp_anom.rolling(time = 3, center = True).mean()
```

```python
def calculate_monthly_deciles(ds, variable_name):
    # Get the shape of the data
    n_months = ds['month'].size
    lat_size = ds['lat'].size
    lon_size = ds['lon'].size
    
    # Create an empty array to hold the decile values
    deciles_array = np.empty((n_months, lat_size, lon_size))
    
    # Loop through each month and calculate deciles
    for i in range(n_months):
        month_data = ds[variable_name].isel(month=i).values
        deciles_array[i, :, :] = calculate_deciles(month_data)
    
    # Create a new DataArray for deciles
    dec = xr.DataArray(
        deciles_array,
        coords=[ds['month'], ds['lat'], ds['lon']],
        dims=['month', 'lat', 'lon']
    )
    
    return dec

ds = sst.sel(time = '2023')
dec = calculate_monthly_deciles(ds, 'sst')

# Add the new dec variable to the original dataset
ds['dec'] = dec
```
