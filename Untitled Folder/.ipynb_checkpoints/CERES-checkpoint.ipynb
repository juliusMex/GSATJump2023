```python
import numpy as np
import xarray as xr
import pandas as pd

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns

from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python

def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

```python
#sst to get oceanic reasons only
fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20,-20), time = slice('2001', '2023'))

#make ocean mask
ocean_mask = ~sst.isel(time = 1).sst.isnull()
```

```python
fp = '/data/jmex/CERES/CERES_TOA.nc'
ceres = xr.open_dataset(fp).where(ocean_mask).sel(time = slice('2001', '2023'))
```

```python
weights = np.cos(np.deg2rad(ceres.lat))
ceres['trp_mean'] = ceres.toa_net_all_mon.weighted(weights).mean(['lat', 'lon'])
ceres['anom'] = ceres.trp_mean.groupby('time.month') - ceres.trp_mean.groupby('time.month').mean()
```

```python
#detrend the data
#compute trend, using subset: 

def detrend_monthly(group):
    ns_century = 1.e9*3600.*24.*365.*100.
    linfit = group.polyfit('time', 1)
    slope = linfit.polyfit_coefficients.isel(degree=0) * ns_century

    trend = xr.polyval(coord=group.time, coeffs=linfit.polyfit_coefficients)
    detrended_group = group - trend.values + group.mean(dim='time')

    return detrended_group


dtr = ceres.anom.groupby('time.month').apply(detrend_monthly)
```

```python
dtr.rolling(time = 6, min_periods =1 , center = True).mean().plot(size = 4, aspect = 4, label = 'Monthly detrended')
dtr.rolling(time = 1, min_periods =1 , center = True).mean().plot(color = 'lightgrey' ,zorder = 1)

#ceres.anom.rolling(center = True, time = 6, min_periods = 1).mean().plot()


plt.title('C; S20N20; 6mth rolling, Ceres')

plt.ylabel('Anom [Wm^-2]')

# Fill between the specified dates
plt.fill_betweenx([-3, 3], np.datetime64('2014-09'), np.datetime64('2015-09'), color='lightgrey', alpha=0.5)
plt.fill_betweenx([-3, 3], np.datetime64('2008-09'), np.datetime64('2009-09'), color='lightgrey', alpha=0.5)
plt.fill_betweenx([-3, 3], np.datetime64('2022-09'), np.datetime64('2023-09'), color='lightgrey', alpha=0.5)

plt.ylim(-2.5, 2.5)
plt.legend()
plt.grid(axis='x')
```

```python
#shortwave TOA; net and clear-sky. downward positive
fp = '/data/jmex/ERA5/monthly/rad_forcing/era5_rad.nc'
R= xr.open_dataset(fp).sel(expver = 1, time = slice('2001', '2023'), latitude = slice(19.5, -19.5))

#change the longitude defintion to 0-360
R['longitude'] = (R['longitude']+360)%360
R = R.sortby('longitude')

#ocean mask
R = R.rename({'longitude':'lon', 'latitude':'lat'})

R = R.sel(lon = slice(110,260)).where(ocean_mask)

#add short and longwave component to net flux
#downwards positive
R['net'] = R.ttr+R.tsr
```

```python
weights = np.cos(np.deg2rad(R.lat))
R['trp_mean'] = R.net.weighted(weights).mean(['lat', 'lon'])
R['anom'] = R.trp_mean.groupby('time.month') - R.trp_mean.groupby('time.month').mean()
```

```python
(R.anom/(3600*24)).plot(size = 4, aspect = 6)
ceres.anom.plot(size = 4, aspect = 6)
```
