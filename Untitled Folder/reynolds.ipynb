```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy.stats import moment
from scipy import stats
import matplotlib.colors as mcolors

from scipy.stats import linregress

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns
from matplotlib.cm import ScalarMappable


from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

## import data

Import data for sst and tp (over ocean)

```python
fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))
```

```python
fp = '/data/jmex/ERA5/monthly/era5_sst24.nc'
sst24 = xr.open_dataset(fp).sel(latitude = slice(20,-20), expver = 1)
#change the longitude defintion to 0-360
sst24['longitude'] = (sst24['longitude']+360)%360
sst24 = sst24.sortby('longitude')
sst24 = sst24.rename({'latitude':'lat', 'longitude':'lon'})

#concat
sst = xr.concat([sst, sst24], dim = 'time')
```

```python
ocean_mask = ~sst.sst.isnull()
weights = np.cos(np.deg2rad(sst.lat))
```

```python
###tp 
fp = '/data/jmex/ERA5/monthly/era5_tp.nc'
tp = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))
```

```python
fp = '/data/jmex/ERA5/monthly/era5_tp24.nc'
tp24 = xr.open_dataset(fp).sel(latitude = slice(20,-20), expver = 1, time = slice('2024-01', '2024-06'))
#change the longitude defintion to 0-360
tp24['longitude'] = (tp24['longitude']+360)%360
tp24 = tp24.sortby('longitude')
tp24 = tp24.rename({'latitude':'lat', 'longitude':'lon'})

tp24['tp'] = tp24.tp*1000

tp = xr.concat([tp, tp24], dim = 'time')
```

```python
tp = tp.where(ocean_mask)
```

```python
# mean precipitation over ocean
tp['tp_mean'] = tp.weighted(weights).mean(['lat', 'lon']).tp
```

```python
# mean precipitation over ocean
sst['sst_mean'] = sst.weighted(weights).mean(['lat', 'lon']).mean
```

```python
#coputing climatologies+anom for sst
clim = sst.sel(time = slice('1990', '2020')).sst.groupby('time.month').mean()

#expand to all time steps
clim_exp = xr.concat([clim.sel(month = t.dt.month) for t in sst.time], dim = 'time')
clim_exp['time'] = sst.time

#add to sst as new variable
sst['clim'] = clim_exp

#calculate anom
sst['anom'] = sst.sst-sst.clim

del clim, clim_exp
```

```python
#coputing climatologies+anom for tp
clim = tp.sel(time = slice('1990', '2020')).tp.groupby('time.month').mean()

#expand to all time steps
clim_exp = xr.concat([clim.sel(month = t.dt.month) for t in tp.time], dim = 'time')
clim_exp['time'] = tp.time

#add to tp as new variable
tp['clim'] = clim_exp

#calculate anom
tp['anom'] = tp.tp-tp.clim

del clim, clim_exp
```
