```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import temperature on diff. pressure levels
fp = '/data/jmex/ERA5/monthly/temp_pressure/era5_t.nc'
t = xr.open_dataset(fp).sel(latitude = slice(20, -20)).sel(expver = 1)
```

```python
# Define constants
p0 = 1000  # reference pressure in hPa
R = 287.05  # specific gas constant for dry air in J/(kg*K)
cp = 1005  # specific heat at constant pressure for dry air in J/(kg*K)

def convert_to_potential_temperature(T, p):
    """
    Convert temperature to potential temperature.
    
    Parameters:
    T (xarray.DataArray): Temperature in K
    p (xarray.DataArray): Pressure in hPa
    
    Returns:
    xarray.DataArray: Potential temperature in K
    """
    return T * (p0 / p) ** (R / cp)

def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

```python
#import sst
fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))

```

```python
weights = np.cos(np.deg2rad(sst.lat))
IPWP = sst.sel(lon = slice(110, 150)).weighted(weights).mean(['lat', 'lon'])
```

```python
#make meridional mean (area weighted)
t = t.weighted(np.cos(np.deg2rad(t.latitude))).mean(['latitude'])
```

```python
#compute potential temperature and its anomaly
t['t_pot'] = convert_to_potential_temperature(t.t, t.level)
t['t_pot_anom'] = t.t_pot.groupby('time.month')-t.t_pot.groupby('time.month').mean()
```

```python
# compute the covariance and variance
cov = (t.t_pot_anom*IPWP.sst).mean(dim = 'year')
var = (IPWp.sst**2).mean()

#slope = covariance/variance
slope = cov/var
```
