```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy.stats import moment
import matplotlib.colors as mcolors
from matplotlib.ticker import MultipleLocator, AutoMinorLocator



import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns

from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

## Import SST

```python
fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))

fp = '/data/jmex/ERA5/monthly/era5_sst24.nc'
sst24 = xr.open_dataset(fp).sel(latitude = slice(20,-20), expver = 1)
#change the longitude defintion to 0-360
sst24['longitude'] = (sst24['longitude']+360)%360
sst24 = sst24.sortby('longitude')
sst24 = sst24.rename({'latitude':'lat', 'longitude':'lon'})

#concat
sst = xr.concat([sst, sst24], dim = 'time')
```

```python
# compute anomalies
# Calculate climatology (1990-2020)
sst_clim = sst.sel(time=slice('1990-01-01', '2020-12-31')).groupby('time.month').mean('time')

# Calculate anomalies
sst['anom'] = sst.sst.groupby('time.month') - sst_clim.sst
```

## Import TP

```python
fp = '/data/jmex/ERA5/monthly/era5_tp.nc'
tp = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))

#in m/month
fp = '/data/jmex/ERA5/monthly/era5_tp24.nc'
tp24 = xr.open_dataset(fp).sel(latitude = slice(20,-20), expver = 1, time = slice('2024-01', '2024-06'))
#change the longitude defintion to 0-360
tp24['longitude'] = (tp24['longitude']+360)%360
tp24 = tp24.sortby('longitude')
tp24 = tp24.rename({'latitude':'lat', 'longitude':'lon'})

tp24['tp'] = tp24.tp*1000

tp = xr.concat([tp, tp24], dim = 'time')
```

```python
# compute anomalies
# Calculate climatology (1990-2020)
tp_clim = tp.sel(time=slice('1990-01-01', '2020-12-31')).groupby('time.month').mean('time')

# Calculate anomalies
tp['anom'] = tp.tp.groupby('time.month') - tp_clim.tp
```

```python
qs = 20
def calculate_deciles(data):
    # Flatten the data to 1D array for qcut
    flattened_data = data.flatten()
    # Calculate deciles using qcut
    deciles, bins = pd.qcut(flattened_data, qs, labels=False, retbins=True)
    # Reshape the result back to the original data shape
    return deciles.reshape(data.shape)
```

```python
def calculate_all_deciles(ds, variable_name):
    # Get the shape of the data
    time_size = ds['time'].size
    lat_size = ds['lat'].size
    lon_size = ds['lon'].size
    
    # Create an empty array to hold the decile values
    deciles_array = np.empty((time_size, lat_size, lon_size))
    
    # Loop through each month and calculate deciles
    for i in range(time_size):
        timestep_data = ds[variable_name].isel(time=i).values
        deciles_array[i, :, :] = calculate_deciles(timestep_data)
    
    # Create a new DataArray for deciles
    dec = xr.DataArray(
        deciles_array,
        coords=[ds['time'], ds['lat'], ds['lon']],
        dims=['time', 'lat', 'lon']
    )
    
    return dec

ds = sst.sel(time = slice('1990', '2023'))
dec = calculate_all_deciles(ds, 'sst')
```

```python
#initalize empty array to store in
tp_per_q = np.zeros((dec.time.size, qs))

#loop thorugh time and quantiles
for it, t in enumerate(dec.time):
    for iq in range(qs):
        #selsct tp_anom in respective quantile
        filtered_data = tp.sel(time = t).where(dec.sel(time = t) == iq, drop = True)
        
        #calculate mean of 'tp' variable for this quantile
        tp_per_q[it, iq] = filtered_data.tp.mean().values
        
        
# safe in a dataarray
tp_qs = xr.DataArray(
    tp_per_q, 
    coords = [dec['time'], np.linspace(100/qs,100,qs)],
    dims = ['time', 'qs'])
```

```python
tanom_qs = tp_qs.groupby('time.month') - tp_qs.groupby('time.month').mean()
```

```python
fig, ax = plt.subplots(1,1, figsize = (20,8))


tanom_qs.T.plot(ax =ax, cmap = 'BrBG', vmin = -2, vmax =2)

```

```python
fig, ax = plt.subplots(1,1, figsize = (5,4))


tanom_qs.sel(time = slice('2023', '2024')).T.plot(ax =ax, cmap = 'BrBG', vmin = -2, vmax = 2)

ax.axvline(np.datetime64('2023-09'))

```
