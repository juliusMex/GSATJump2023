```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy.stats import moment
import matplotlib.colors as mcolors
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

from scipy.stats import linregress

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns

from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

```python
# import SST
fp = '/data/jmex/ERA5/monthly/era5_sst.nc'
sst = decode_time(fp).sel(lat = slice(20,-20), time = slice('1990', '2023'))

ocean_mask = ~sst.isel(time =1).sst.isnull()
```

```python
#take ASO only and make season average
sst = sst.where(sst.time.dt.month.isin([8,9,10])).groupby('time.year').mean()

#compute anom
sst = sst - sst.mean('year')
```

```python
# make nino34 time series
#select region
nino34 = sst.sel(lat = slice(5, -5), lon = slice(190, 240))

#weight and take average
nino34 = nino34.weighted(np.cos(np.deg2rad(nino34.lat))).mean(['lat', 'lon'])
```

## SST

```python
# compute the covariance and variance
cov = (sst*nino34).mean(dim = 'year')
var = (nino34**2).mean()

#slope = covariance/variance
slope = cov/var
```

```python
slope.sst.plot(size =4, aspect =6)
```

```python
# compute the nino34 for 2023 ASO
#multiply in with the regression pattern to compute the 'counterfactual'

counterfactual23 = nino34.sel(year = 2023)*slope
```

```python
# compute the nino34 for 2015 ASO
#multiply in with the regression pattern to compute the 'counterfactual'

counterfactual15 = nino34.sel(year = 2015)*slope
```

```python
# compute the nino34 for 2009 ASO
#multiply in with the regression pattern to compute the 'counterfactual'

counterfactual09 = nino34.sel(year = 2009)*slope
```

```python
# Make the plots of 2023 ASO SST; the counterfactual; the difference
#real
sst.sel(year = 2023).sst.plot(size  =4 , aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Observation 2023')
#conterfactual
counterfactual23.sst.plot(size = 4, aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Counterfactual 2023')
#difference
(sst.sel(year = 2023).sst-counterfactual23.sst).plot(size = 4, aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Observation - Counterfactual 2023');
```

```python
# Make the plots of 2015 ASO SST; the counterfactual; the difference
#real
sst.sel(year = 2015).sst.plot(size  =4 , aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Observation 2015')
#conterfactual
counterfactual15.sst.plot(size = 4, aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Counterfactual 2015')
#difference
(sst.sel(year = 2015).sst-counterfactual15.sst).plot(size = 4, aspect = 6, vmin = -3, vmax = 3, cmap = 'RdBu_r')
plt.title('Observation - Counterfactual 2015');
```
