```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy.stats import moment
from scipy import stats
import matplotlib.colors as mcolors

from scipy.stats import linregress

import matplotlib.pyplot as plt
import calendar
import matplotlib.patches as mpatches
import seaborn as sns
from matplotlib.cm import ScalarMappable


from pylab import rcParams
rcParams['axes.labelweight'] = 'bold'
rcParams['axes.labelsize'] = 'x-large'
rcParams['axes.titlesize'] = 'xx-large'
rcParams['axes.titleweight'] = 'bold'
```

```python
fp  = '/data/jmex/ERA5/monthly/clouds/era5_clouds.nc'
clouds = xr.open_dataset(fp)
```

```python
clouds = clouds.drop(['expver', 'number'])
clouds = clouds.rename({'latitude':'lat', 'longitude':'lon', 'valid_time':'time'})
clouds['lon'] = (clouds['lon']+360)%360
```

```python
#cmmpute means
weights = np.cos(np.deg2rad(clouds.lat))
clouds_mean = clouds.weighted(weights).mean(['lat', 'lon']) 
```

```python
clouds_frac = clouds_mean/clouds_mean.tcc
```

```python
clim = clouds_frac.sel(time = slice('1991', '2020')).groupby('time.month').mean()
anom = clouds_frac.groupby('time.month')-clim
```

```python
anom = anom.rolling(time = 3, center = True, min_periods = 1).mean()
```

```python
fig, axes = plt.subplots(3,1, figsize = (12,12))

anom.lcc.plot(ax = axes[0])
anom.mcc.plot(ax = axes[1])
anom.hcc.plot(ax = axes[2])
```

```python
r = clouds_mean.lcc/clouds_mean.tcc
r = r.groupby('time.month') - r.sel(time = slice('1991', '2020')).groupby('time.month').mean()
r = r.where(r.time.dt.month.isin([1,2,3,4,5,6, 7, 8, 9])).groupby('time.year').mean()

r_ = r - r.roll(year = 1)
r_.plot(size = 6, aspect = 4)
plt.xticks(r_.year[::2]);
plt.grid()
```

```python
r = clouds_mean.lcc/clouds_mean.tcc
r = r.groupby('time.month') - r.sel(time = slice('1991', '2020')).groupby('time.month').mean()
r = r.groupby('time.year').mean()

r_ = r - r.roll(year = 1)
r_.plot(size = 6, aspect = 4)
plt.xticks(r_.year[::2]);
plt.grid()
```

```python
fig, axes = plt.subplots(3,1, figsize = (4,12))

clouds_frac.lcc.plot(ax = axes[0])
clouds_frac.mcc.plot(ax = axes[1])
clouds_frac.hcc.plot(ax = axes[2])
```

```python
def decode_time(filepath):
    ds = xr.open_dataset(filepath, decode_times=False)
    units, reference_date = ds.time.attrs['units'].split('since')
    ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')
    return ds
```

```python
#load the masks
fp = '/data/jmex/masks/RECCAP2_region_masks_all.nc'
masks = xr.open_dataset(fp).sel(lat = slice(-15,25))
ocean_mask = masks.open_ocean.isin([2,3]) # mask for tropical indo-pacifi
```

```python
#clouds
fp = '/data/jmex/ERA5/monthly/rad_forcing/era5_clouds.nc'
clouds = xr.open_dataset(fp).sel(expver = 1, latitude = slice(20,-20), time = slice('1990', '2023'))
#change the longitude defintion to 0-360
clouds['longitude'] = (clouds['longitude']+360)%360
clouds = clouds.sortby('longitude')

clouds = clouds.rename({'latitude':'lat', 'longitude':'lon'})
```

```python
clouds
```

```python
clouds = clouds.where(ocean_mask)
```

```python
weights = np.cos(np.deg2rad(clouds.lat))

clouds_mean = clouds.weighted(weights).mean(['lat', 'lon'])
```

```python
clim = clouds_mean.sel(time = slice('2000', '2020')).groupby('time.month').mean()

clouds_anom = clouds_mean.groupby('time.month') - clim
```

```python
r = clouds_mean.lcc/clouds_mean.hcc
r_anom = r.groupby('time.month') - r.groupby('time.month').mean()
```

```python
r_anom.plot(size = 4, aspect = 4)
plt.xticks(r_anom.time[::12])
plt.grid()
```

```python
r = clouds.lcc/clouds.hcc
```

```python
r = r.groupby('time.month') - r.groupby('time.month').mean()
```

```python
(r.sel(time  = slice('2023-04', '2023-09')).mean('time')-r.sel(time  = slice('2022-04', '2022-09')).mean('time')).plot(size = 4, aspect = 7, 
                                        vmin = -1, vmax = 1)
```
